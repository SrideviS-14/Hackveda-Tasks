```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'path_to_csv_file' with the actual path to your CSV file.
df = pd.read_csv("C:\\Users\\91984\\Desktop\\Internships\\Hackveda\\Plotting trend lines on HDFCBANK CSV Dataset\\StockDataHDFCBANK.csv")

# Convert the 'Date' column to a datetime data type and set it as the index.
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Closing Price', color='blue')
plt.title('HDFCBANK Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
```

```python
# Calculate moving averages for higher lows and lower highs
df['20MA'] = df['Close'].rolling(window=20).mean()
df['50MA'] = df['Close'].rolling(window=50).mean()

# Plot the trend lines
plt.plot(df.index, df['20MA'], label='20-Day MA', color='red')
plt.plot(df.index, df['50MA'], label='50-Day MA', color='green')
```

```python
# Add parallel lines for uptrend
plt.axhline(df['20MA'].max(), color='purple', linestyle='--', label='Uptrend Line')
plt.annotate('Uptrend', (df.index[100], df['20MA'].max()), textcoords="offset points", xytext=(0,10), ha='center')

# Add parallel lines for downtrend
plt.axhline(df['50MA'].min(), color='orange', linestyle='--', label='Downtrend Line')
plt.annotate('Downtrend', (df.index[150], df['50MA'].min()), textcoords="offset points", xytext=(0,10), ha='center')
```
